## Problem 2

```python
# same intrisincs for both cameras
K = np.array([[1.18468085e+04, 0.00000000e+00, 2.78400000e+03],
              [0.00000000e+00, 1.18216561e+04, 1.85600000e+03],
              [0.00000000e+00, 0.00000000e+00, 1.00000000e+00]])
# extrinsincs for camera 1
E1 = np.array([[-0.31402103, -0.94879008,  0.03446983, -0.94640895],
               [ 0.13790285, -0.08150285, -0.98708667,  1.42010489],
               [ 0.93934743, -0.30521248,  0.15643447,  0.85907637]])
# extrinsics for camera 2
E2 = np.array([[-0.6022098 , -0.79814732, -0.01744177,  0.54587954],
               [-0.01736461,  0.03493795, -0.99923861,  1.4551765 ],
               [ 0.798149  , -0.60144841, -0.0348995 ,  0.75675417]])
# estimated position of world point (meters, in world coords)
Pw = np.array([[ 4.3 ],
               [-1.4 ],
               [ 2.03]])
# observed location of Pw in each camera (col,row pixel units)
px_cam1 = np.array([[ 898.91171569],
                    [2114.24202796]])
px_cam2 = np.array([[488.47708728],
                    [198.14597679]])
```

Create a function to get both cameras projection point.

```python
def project_point(point_3d, K, E):
    point_3d_homogeneous = np.append(point_3d, 1)
    point_2d_homogeneous = np.dot(E, point_3d_homogeneous)
    pixel_2d_homogeneous = np.dot(K, point_2d_homogeneous)
    pixel_2d = pixel_2d_homogeneous[:2] / pixel_2d_homogeneous[2]
    return np.array([[pixel_2d[0]], [pixel_2d[1]]])
```

```python
projected_cam1 = project_point(Pw, K, E1)
projected_cam2 = project_point(Pw, K, E2)
```

Calculate and print out the squared error for each camera.

```python
cam1_error = projected_cam1 - px_cam1
cam2_error = projected_cam2 - px_cam2
squared_cam1_error = np.sum((cam1_error)**2)
squared_cam2_error = np.sum((cam2_error)**2)

print(squared_cam1_error)
print(squared_cam2_error)
```
